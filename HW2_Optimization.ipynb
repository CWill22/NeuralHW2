Perform the optimization problem of finding the minimum of $J(\theta) = (2\theta - 3)^2$ by:

1. Plotting $J(\theta)$ vs. $\theta$.
2. Determining its minimum using the gradient descent approach starting from a random initial value of $\theta = 5$. Perform the search for the minimum using the gradient descent approach by hand calculations, i.e., step 1, step 2, etc., showing your work completely.


Set up libraries and equations

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Define the objective function J(theta) and its derivative dJ/dtheta
def J(theta):
    """Compute the objective function J(theta) = (2*theta - 3)^2."""
    return (2 * theta - 3)**2

def dJ(theta):
    """Compute the derivative of J with respect to theta.

    J(theta) = (2*theta - 3)^2
    => dJ/dtheta = 2*(2*theta - 3)*2 = 4*(2*theta - 3)
    """
    return 4 * (2 * theta - 3)
```

1. Plotting $J(\theta)$ vs. $\theta$.



```python
# Create an array of theta values for plotting
theta_vals = np.linspace(-1, 4, 100)
J_vals = J(theta_vals)

# Plot the function J(theta)
plt.figure(figsize=(8, 6))
plt.plot(theta_vals, J_vals, label=r'$J(\theta) = (2\theta-3)^2$', linewidth=2)
plt.xlabel(r'$\theta$', fontsize=14)
plt.ylabel(r'$J(\theta)$', fontsize=14)
plt.title('Plot of $J(\\theta) = (2\\theta-3)^2$', fontsize=16)
plt.legend()
plt.grid(True)
plt.show()
```

By graphing, we can see the minimum is at the bottom of the parabala, and is (1.5, 0)

2. Determining its minimum using the gradient descent approach starting from a random initial value of $\theta = 5$. Perform the search for the minimum using the gradient descent approach by hand calculations, i.e., step 1, step 2, etc., showing your work completely.

```python
# Initialize parameters for gradient descent
theta = 5.0         # starting point
alpha = 0.1         # learning rate
num_iterations = 10 # number of iterations to perform

print("Iteration\t theta\t\t J(theta)")
for i in range(num_iterations):
    current_J = J(theta)
    print(f"{i:2d}\t\t {theta:.4f}\t {current_J:.4f}")

    # Compute the gradient at the current theta
    grad = dJ(theta)

    # Update theta: theta_new = theta_old - alpha * (gradient)
    theta = theta - alpha * grad

# Print the final value after the iterations
print(f"Final value after {num_iterations} iterations: theta = {theta:.4f}, J(theta) = {J(theta):.4f}")
```

```python
# Reinitialize the starting point and store the theta values for each iteration
theta = 5.0
theta_history = [theta]

for _ in range(num_iterations):
    theta = theta - alpha * dJ(theta)
    theta_history.append(theta)

theta_history = np.array(theta_history)
J_history = J(theta_history)

# Plot the function and the gradient descent steps
plt.figure(figsize=(8, 6))
plt.plot(theta_vals, J_vals, label=r'$J(\theta)$', linewidth=2)
plt.plot(theta_history, J_history, 'ro-', label='Gradient Descent Steps', markersize=6)
plt.xlabel(r'$\theta$', fontsize=14)
plt.ylabel(r'$J(\theta)$', fontsize=14)
plt.title('Gradient Descent Path on $J(\\theta)$', fontsize=16)
plt.legend()
plt.grid(True)
plt.show()
```
